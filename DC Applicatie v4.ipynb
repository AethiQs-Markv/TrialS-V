```python
#Ingelezen sheets bewerken naar percentages uit lifecycles en duraties
aanbieders=np.unique(Lifecycles.index.values)
kolomwaarden=['Staats kort','Staats lang','Credits','High Yield','EMD','Ontwikkeld','Opkomend', 'Vastgoed', 'Commodities']
duratienamen=['Duratie staats kort','Duratie staats lang','Duratie credits']

Percentages={}
Duraties={}
for partij in aanbieders:
    for waarde in kolomwaarden:
        Percentages[partij,waarde]=Lifecycles.loc[partij,['Leeftijd',waarde]]
        Percentages[partij,waarde].set_index('Leeftijd',inplace=True)
    for naam in duratienamen:
        Duraties[partij,naam]=Lifecycles.loc[partij,['Leeftijd',naam]]
        Duraties[partij,naam].set_index('Leeftijd',inplace=True)
```

```python
##Deze functie is voor het berekenen van kapitalen per aanbieder
def doorrekening(leeftijd,salaris,startkapitaal,aanbieder,staffelkeuze):
    
    kapitaal=np.zeros((1000,pensioenrichtleeftijd-leeftijd+1))
    Rendement=np.zeros((1000,pensioenrichtleeftijd-leeftijd))
    
    #berekening van DC inleg afhankelijk van staffelkeuze en leeftijd
    PG0=max(salaris-franchise,0) 
    tile=np.tile(Staffels.loc[leeftijd:(pensioenrichtleeftijd-1),staffelkeuze],(1000,1))
    Inleg=np.multiply(PG0,np.multiply(CumulatiefLooninflatie.loc[:,0:(pensioenrichtleeftijd-leeftijd-1)],tile))
    
    #Berekening van rendement o.b.v. de verschillende categorieeën
    for categorie in Namencategorieen:
        T=np.tile(np.transpose(Percentages[(aanbieder,categorie)].loc[leeftijd:(pensioenrichtleeftijd-1)]),(1000,1))
        hulp=np.multiply(Rendementen[categorie].loc[:,0:(pensioenrichtleeftijd-leeftijd-1)],T)
        Rendement=np.add(Rendement,hulp)
        
    #Kapitaal ontwikkeling
    for i in range(0, 1000):    
        kapitaal[i,0]=startkapitaal
        
        for j in range(0,pensioenrichtleeftijd-leeftijd):
            kapitaal[i,j+1]=kapitaal[i,j]*Rendement.loc[i,j]+Inleg.loc[i,j]  
    
    return kapitaal
```

```python
#Formule om een boxplot te genereren 
def boxplot(dictionary):

    df=pd.DataFrame.from_dict(dictionary)
    s=df.stack()
    df=pd.DataFrame(s,index=None)
    df.reset_index(inplace=True)
    df.drop(['level_0'],axis=1,inplace=True)
    df.columns=['Aanbieders','Kapitaal']

    
    boxwhisker = hv.BoxWhisker(df, 'Aanbieders', 'Kapitaal')
    boxwhisker=boxwhisker(plot={'width':500, 'height':500,'title_format':'Hoeveel pensioen heeft u bij iedere aanbieder op de pensioendatum?','xrotation':45})
    #%%opts BoxWhisker [height=500 width=500 xrotation=45]
    return boxwhisker
```

```python
def ontwikkelingkapitaalgrafiek(kapitaal,leeftijd):     
    ##Berekenen percentielen over ieder jaar
        perc5 = np.percentile(kapitaal,5,axis=0)
        perc25 = np.percentile(kapitaal,25,axis=0)
        perc50 = np.percentile(kapitaal,50,axis=0)
        perc75 = np.percentile(kapitaal,75,axis=0)
        perc95 = np.percentile(kapitaal,95,axis=0)

        ##Ten behoeve van stacked onderdelen
        perc25_5 = np.around(np.subtract(perc25, perc5), decimals=0)
        perc50_25 = np.around(np.subtract(perc50, perc25), decimals=0)
        perc75_50 = np.around(np.subtract(perc75, perc50), decimals=0)
        perc95_75= np.around(np.subtract(perc95, perc75), decimals=0)

        lft=list(range(leeftijd,pensioenrichtleeftijd+1))

        ##Opzet Area Chart
        dims = dict(kdims='Leeftijd', vdims=[hv.Dimension('Kapitaal (€)')])
        grafiek1 = hv.Area((lft,perc5), **dims)
        grafiek2  = hv.Area((lft,perc25_5), **dims)
        grafiek3 = hv.Area((lft,perc50_25),**dims)
        grafiek4= hv.Area((lft,perc75_50),**dims)
        grafiek5 = hv.Area((lft,perc95_75),**dims)

        ##Aanpassen Kleurstelling naar blauw en samenvoegen grafieken
        grafiek1 = grafiek1(style={'fill_color':'White'},plot={'width':500, 'height':500, 'tools':['hover'],'title_format':'Ontwikkeling pensioenkapitaal'})
        grafiek2 = grafiek2(style={'fill_color':Blues4[1]},plot={'tools':['hover']})
        grafiek3 = grafiek3(style={'fill_color':Blues4[0]},plot={'tools':['hover']})
        grafiek4 = grafiek4(style={'fill_color':Blues4[0]},plot={'tools':['hover']})
        grafiek5 = grafiek5(style={'fill_color':Blues4[1]},plot={'tools':['hover']})

        overlay=(grafiek1 * grafiek2 *grafiek3 * grafiek4 * grafiek5)
        return hv.Area.stack(overlay).relabel('Grafiek') 
```

```python
##Hoofdfunctie om output te genereren
class parameters(hv.streams.Stream):
    leeftijd = param.Number(default=25, bounds=(25, 66))
    salaris = param.Number(default=1000, bounds=(0, 103317))
    startkapitaal = param.Number(default=1200, bounds=(1000, 200000))
    aanbieder = param.ObjectSelector(default=aanbieders[0], objects=aanbieders)
    staffelkeuze = param.ObjectSelector(default='Staffel - 3% rekenrente', objects=['Staffel - 3% rekenrente','Staffel - 4% rekenrente'])
    
    def output(self,**kwargs):

        #Initialiseren dictionary om de kapitaalontwikkeling van alle aanbieders in op te slaan
        kapitaal={}
        eindkapitaal={}

        #doorrekenen alle aanbieders
        for hulp in aanbieders:
            kapitaal[hulp]=doorrekening(self.leeftijd,self.salaris,self.startkapitaal,hulp,self.staffelkeuze)
            eindkapitaal[hulp]=pd.Series(kapitaal[hulp][:,-1])
        
        grafiek1=ontwikkelingkapitaalgrafiek(kapitaal[self.aanbieder],self.leeftijd) 
        grafiek2=boxplot(eindkapitaal) 
        samen=grafiek1 + grafiek2
        samen=samen(plot={'title_format':'Ontwikkeling pensioenkapitaal'})

        return samen
    
    
vergelijker = parameters(name='Parameters')   
dmap=hv.DynamicMap(vergelijker.output, streams=[vergelijker])
plot=hv.renderer('bokeh').get_plot(dmap,doc=Document())
parambokeh.Widgets(vergelijker, view_position='right',callback=vergelijker.event,plots=[plot])
```
